```python
df_country=pd.read_excel("C:\\Users\\shirisha\\Downloads\\Country-Code.xlsx")
df_country.head()
```

```python
df_final=pd.merge(df,df_country,on='Country Code',how='left')
df_final.head()
```

```python
#finding country
country_name=df_final.Country.value_counts().index
country_name
```

```python
country_values=df_final.Country.value_counts().values
country_values
```

```python
#creating pechart
plt.pie(country_values,labels=country_name)
```

```python
#finding t0p 4 country
plt.pie(country_values[:3],labels=country_name[:3],autopct='%1.2f%%')
```

```python
df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
ratings=df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating'})
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating",hue="Rating color",data=ratings,palette=["grey","green","Yellow","orange","blue","black"])
```
